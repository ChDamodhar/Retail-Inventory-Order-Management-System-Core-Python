Data Correlations- Finding Relationships

A great aspect of the Pandas module is the corr() method.

The corr() method calculates the relationship between each column in your data set.

df.corr()

Note: The corr() method ignores "not numeric" columns.

                      Duration     Pulse         Maxpulse    Calories

  Duration  1.000000    -0.155408  0.009403      0.922721

  Pulse        -0.155408    1.000000   0.786535      0.025120

  Maxpulse  0.009403    0.786535   1.000000      0.203814

  Calories  0.922721    0.025120   0.203814      1.000000

The Result of the corr() method is a table with a lot of numbers that represents how well the relationship is between two columns.

The number varies from -1 to 1.

1 means that there is a 1 to 1 relationship (a perfect correlation), and for this data set, each time a value went up in the first column, the other one went up as well.

0.9 is also a good relationship, and if you increase one value, the other will probably increase as well.

-0.9 would be just as good relationship as 0.9, but if you increase one value, the other will probably go down.

0.2 means NOT a good relationship, meaning that if one value goes up does not mean that the other will.

What is a good correlation? It depends on the use, but I think it is safe to say you have to have at least 0.6 (or -0.6) to call it a good correlation.

Perfect Correlation:

We can see that "Duration" and "Duration" got the number 1.000000, which makes sense, each column always has a perfect relationship with itself.

Good Correlation:

"Duration" and "Calories" got a 0.922721 correlation, which is a very good correlation, and we can predict that the longer you work out, the more calories you burn, and the other way around: if you burned a lot of calories, you probably had a long work out.

Bad Correlation:

"Duration" and "Maxpulse" got a 0.009403 correlation, which is a very bad correlation, meaning that we can not predict the max pulse by just looking at the duration of the work out, and vice versa.

```python
import pandas as pd

path = '/content/drive/My Drive/OnlineRetail.csv'
df1 = pd.read_csv(path, encoding='latin1')

df1.columns = df1.columns.str.strip()

print(df1.head())
print(df1.tail())
df1.info()
print(df1.shape)
print(df1.dtypes)

print(df1.isnull().sum())
df1 = df1.dropna(subset=['CustomerID'])
df1['Description'] = df1['Description'].fillna('Unknown')

df1['InvoiceDate'] = pd.to_datetime(df1['InvoiceDate'], errors='coerce')
df1 = df1.dropna(subset=['InvoiceDate'])

df1 = df1[df1['Quantity'] > 0]
df1 = df1[df1['UnitPrice'] > 0]

print(df1.duplicated().sum())
df1 = df1.drop_duplicates()

df1.info()
print(df1.shape)
print(df1.head())

df1['TotalPrice'] = df1['Quantity'] * df1['UnitPrice']
df1['Year'] = df1['InvoiceDate'].dt.year
df1['Month'] = df1['InvoiceDate'].dt.month


high_quantity = df1[df1['Quantity'] > 1000]
print(high_quantity)

df1['Revenue'] = df1['Quantity'] * df1['UnitPrice']
sorted_df = df1.sort_values(by='Revenue', ascending=False)
print(sorted_df.head())

product_info = pd.DataFrame({
    'StockCode': ['85123A', '71053', '84406B', '84029G', '84029E'],
    'Category': ['Home Decor', 'Home Decor', 'Home Decor', 'Home Decor', 'Home Decor'],
    'Supplier': ['SupplierA', 'SupplierB', 'SupplierC', 'SupplierD', 'SupplierE']
})

df1 = pd.merge(df1, product_info, on='StockCode', how='left')
print(df1.head())

country_stats = df1.groupby('Country').agg({
    'Revenue': ['sum', 'mean'],
    'InvoiceNo': 'count'
}).rename(columns={'InvoiceNo':'OrderCount'})
print(country_stats)

bins = [0, 5, 20, df1['UnitPrice'].max()]
labels = ['Cheap', 'Medium', 'Expensive']
df1['PriceRange'] = pd.cut(df1['UnitPrice'], bins=bins, labels=labels)
price_stats = df1.groupby('PriceRange').agg({
    'Revenue':'mean',
    'InvoiceNo':'count'
}).rename(columns={'InvoiceNo':'PurchaseCount'})
print(price_stats)

numeric_corr = df1[['Quantity', 'UnitPrice', 'Revenue']].corr()
print(numeric_corr)

df1 = df1.rename(columns={'StockCode':'Product_ID', 'UnitPrice':'Price'})
df1['Description_lower'] = df1['Description'].str.lower()
glass_products = df1[df1['Description_lower'].str.contains('glass', na=False)]
print(glass_products.head())



numeric_stats = df1.describe()
print(numeric_stats)

country_counts = df1['Country'].value_counts()
print(country_counts)

category_counts = df1['Category'].value_counts(dropna=False)
print(category_counts)

high_revenue_germany = df1.query('Revenue > 1000 and Country == "Germany"')
print(high_revenue_germany.head())

quantity_range = df1[df1['Quantity'].between(10, 100)]
print(quantity_range.head())

df1.to_csv('/content/drive/My Drive/Cleaned_OnlineRetail.csv', index=False)
df1.to_excel('/content/drive/My Drive/Cleaned_OnlineRetail.xlsx', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress

path = '/content/drive/My Drive/data.csv'
df = pd.read_csv(path)
df = df.dropna(subset=['Duration', 'Calories'])

x = df['Duration']
y = df['Calories']
correlation = x.corr(y)
slope, intercept, _, _, _ = linregress(x, y)

plt.figure(figsize=(8,5))
plt.scatter(x, y, color='purple', label=f'Pearson r = {correlation:.2f}')
plt.plot(x, slope*x + intercept, color='red', linestyle='--', label='Trend line')
plt.xlabel("Duration")
plt.ylabel("Calories")
plt.title("Duration vs Calories")
plt.legend()
plt.tight_layout()
plt.savefig("/content/scatter_duration_vs_calories.png")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

path = '/content/drive/My Drive/data.csv'
df = pd.read_csv(path)
df = df.dropna(subset=['Maxpulse'])

ranges = ['<=120', '121-140', '141-160', '>160']
counts = [
    (df['Maxpulse'] <= 120).sum(),
    ((df['Maxpulse'] > 120) & (df['Maxpulse'] <= 140)).sum(),
    ((df['Maxpulse'] > 140) & (df['Maxpulse'] <= 160)).sum(),
    (df['Maxpulse'] > 160).sum()
]

colors = ['skyblue', 'lightgreen', 'orange', 'pink']
explode = [0.0, 0.05, 0.0, 0.0]

plt.figure(figsize=(6,6))
plt.pie(counts, labels=ranges, autopct='%1.1f%%', colors=colors, explode=explode, shadow=True, startangle=90)
plt.title("Maxpulse Range Distribution")
plt.tight_layout()
plt.savefig("/content/pie_maxpulse_ranges.png")
plt.show()
```

```python
from google.colab import drive
import pandas as pd

drive.mount('/content/drive')
df = pd.read_csv('/content/drive/My Drive/Colab Notebooks/Bengaluru_House_Data.csv')

print(df.head())
print(df.tail())
print(df.describe())     # summary statistics
print(df.info())
print(df.columns)
```

```python
print(df.isnull().sum())
df.drop(['society'],axis=1,inplace=True,errors='ignore')
df['bath'].fillna(df['bath'].median(),inplace=True)
df['balcony'].fillna(0,inplace=True)
# total_sqft cleanup
def convert_sqft(x):
  if isinstance(x,str) and '-' in x:
    a,b = x.split('-')
    return (float(a)+float(b))/2
  try:
    return float(x)
  except:
    return None

df['total_sqft']=df['total_sqft'].apply(convert_sqft)
df.dropna(subset=['total_sqft'],inplace=True)
df.drop_duplicates(inplace=True)
df.reset_index(drop=True,inplace=True)
df.head()
```

```python
unique_locations = df['location'].nunique()
print(unique_locations)

avg_price_by_location = df.groupby('location')['price'].mean().sort_values()
print(avg_price_by_location)

highest_avg_price_location = avg_price_by_location.idxmax()
highest_avg_price = avg_price_by_location.max()

print(highest_avg_price_location, highest_avg_price)

correlation_values = df[['total_sqft','bath','price']].corr()
correlation_values

corr_sqft_price = correlation_values.loc['total_sqft', 'price']
print("\n5) Interpretation:")

if corr_sqft_price > 0.7:
    print("â€¢ Strong positive correlation â†’ Larger houses generally have higher prices.")
elif corr_sqft_price > 0.4:
    print("â€¢ Moderate correlation â†’ Larger houses often cost more, but not always.")
else:
    print("â€¢ Weak correlation â†’ Size alone does NOT determine price.")



```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set clean style
sns.set(style="whitegrid", palette="deep")
plt.figure(figsize=(6,4))


# 1. Price Distribution (Histogram + KDE)
plt.figure(figsize=(6,4))
sns.histplot(df['price'], kde=True)
plt.title("Price Distribution", fontsize=16)
plt.xlabel("Price (Lakhs or Crores)")
plt.ylabel("Frequency")
plt.show()


# 2. Relationship Between Area and Price â€” Scatter Plot
plt.figure(figsize=(6,4))
sns.scatterplot(x=df['total_sqft'], y=df['price'])
plt.title("Total Sqft vs Price", fontsize=16)
plt.xlabel("Total Sqft")
plt.ylabel("Price")
plt.show()


# 3. Effect of Bathrooms on Price â€” Box Plot
plt.figure(figsize=(6,4))
sns.boxplot(x=df['bath'], y=df['price'])
plt.title("Effect of Bathroom Count on Price", fontsize=16)
plt.xlabel("Number of Bathrooms")
plt.ylabel("Price")
plt.show()


# 4. Top 10 Most Expensive Locations â€” Bar Chart
avg_price_by_location = df.groupby('location')['price'].mean().sort_values(ascending=False)
top_10_locations = avg_price_by_location.head(10)

plt.figure(figsize=(6,4))
sns.barplot(x=top_10_locations.index, y=top_10_locations.values)
plt.xticks(rotation=45, ha='right')
plt.title("Top 10 Locations with Highest Average House Price", fontsize=16)
plt.xlabel("Location")
plt.ylabel("Average Price")
plt.tight_layout()
plt.show()


# 5. Correlation Heatmap
plt.figure(figsize=(6,4))
numeric_df = df.select_dtypes(include=['int64','float64'])
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Numeric Features", fontsize=16)
plt.show()
```

```python
from google.colab import drive
import pandas as pd

drive.mount('/content/drive')
df = pd.read_csv('/content/drive/My Drive/Colab Notebooks/amazon_products_dataset.csv')

print( df.head())
print(df.tail())
print(df.info())
print(df.shape)
print(df['category'].unique())

print("\nCOLUMNS WITH MISSING VALUES:")
print(df.isnull().sum())

print("\nDATA TYPES OF COLUMNS:")
print(df.dtypes)



#task2
print(df.isnull().sum())

numeric_cols = ['discount_price', 'actual_price', 'rating', 'rating_count']
for col in numeric_cols:
    if col in df.columns:
        df[col] = df[col].fillna(df[col].median())

categorical_cols = ['brand', 'category']
for col in categorical_cols:
    if col in df.columns:
        df[col] = df[col].fillna("Unknown")

df = df.drop_duplicates()

for col in numeric_cols:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')

if 'rating_count' in df.columns:
    df['rating_count'] = df['rating_count'].astype('Int64')

df = df.reset_index(drop=True)

df.head()


#Task3
# Top 5 expensive products
top_5_expensive = df.sort_values(by="actual_price", ascending=False).head(5)
print(top_5_expensive)

# Brand with most products
top_brand = df['brand'].value_counts().idxmax()
top_brand_count = df['brand'].value_counts().max()
print("Brand with highest products:", top_brand)
print("Number of products:", top_brand_count)

print("Calculate discount percentage")
df['discount_percentage'] = ((df['actual_price'] - df['discount_price']) / df['actual_price']) * 100

print("Average discount per brand")
avg_discount_per_brand = df.groupby('brand')['discount_percentage'].mean().sort_values(ascending=False)
print(avg_discount_per_brand)

print("High rating products")
high_rating_products = df[(df['rating'] >= 4.5) & (df['rating_count'] >= 2000)]
print(high_rating_products)

print("Category-wise average price")
category_avg_price = df.groupby('category')['actual_price'].mean().sort_values(ascending=False)
print(category_avg_price)



```

task-4a

```python
import numpy as np

df['discount_percent'] = ((df['actual_price'] - df['discount_price']) / df['actual_price']) * 100

def categorize_price(price):
    if price < 500:
        return "Budget"
    elif 500 <= price < 2000:
        return "Midrange"
    else:
        return "Premium"

df['price_category'] = df['actual_price'].apply(categorize_price)
df['popularity_score'] = df['rating'] * np.log(df['rating_count'] + 1)
print(df[['actual_price', 'discount_price', 'discount_percent', 'price_category', 'rating', 'rating_count', 'popularity_score']].head())
```

task-4b

```python
top_10_popularity = df.sort_values(by='popularity_score', ascending=False).head(10)
print(top_10_popularity)

premium_high_discount = df[(df['price_category'] == 'Premium') & (df['discount_percent'] > 40)]
print(premium_high_discount)

bluetooth_products = df[df['title'].str.contains('Bluetooth', case=False, na=False)]
print(bluetooth_products)

popularity_threshold = df['popularity_score'].quantile(0.7)
filtered_products = df[
    (df['rating'] >= 4.0) &
    (df['discount_percent'].between(20, 50)) &
    (df['popularity_score'] >= popularity_threshold)
]
print(filtered_products)
```

task-4c

```python
avg_discount_per_brand = df.groupby('brand')['discount_percent'].mean().sort_values(ascending=False)
print(avg_discount_per_brand)

avg_popularity_per_category = df.groupby('price_category')['popularity_score'].mean().sort_values(ascending=False)
print(avg_popularity_per_category)

highest_discount_category = df.groupby('price_category')['discount_percent'].mean().idxmax()
print(highest_discount_category)

df['revenue_potential'] = df['actual_price'] * df['rating_count']
total_revenue_per_brand = df.groupby('brand')['revenue_potential'].sum().sort_values(ascending=False)
print(total_revenue_per_brand)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.histplot(df['actual_price'], kde=True, bins=50, color='skyblue')
plt.title('Price Distribution')
plt.xlabel('Actual Price')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(10,6))
sns.scatterplot(x='actual_price', y='rating', data=df, hue='price_category', alpha=0.6)
plt.title('Price vs Rating')
plt.xlabel('Actual Price')
plt.ylabel('Rating')
plt.show()

category_avg_rating = df.groupby('category')['rating'].mean().sort_values(ascending=False)
plt.figure(figsize=(12,6))
sns.barplot(x=category_avg_rating.index, y=category_avg_rating.values, palette='viridis')
plt.xticks(rotation=90)
plt.title('Category-wise Average Rating')
plt.xlabel('Category')
plt.ylabel('Average Rating')
plt.show()

plt.figure(figsize=(12,6))
sns.countplot(y='brand', data=df, order=df['brand'].value_counts().index, palette='magma')
plt.title('Brand Product Count')
plt.xlabel('Count')
plt.ylabel('Brand')
plt.show()

plt.figure(figsize=(10,8))
numeric_cols = ['actual_price', 'discount_price', 'discount_percent', 'rating', 'rating_count', 'popularity_score']
corr = df[numeric_cols].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Numeric Feature Correlations')
plt.show()
```
